## 2. Functional vs. Non-Functional Testing

| Type | Description | Examples |
| :--- | :--- | :--- |
| **Functional** | Checks *what* the system does | Login, Button clicks, Data saving |
| **Non-Functional** | Checks *how* the system performs | Speed, Security, Usability, Load |

### Scenarios Classification

| Scenario | Type |
| :--- | :--- |
| User logs in with valid credentials | Functional |
| Page loads within 2 seconds | Non-Functional |
| Password is encrypted | Non-Functional |
| Add item to cart works | Functional |
| System handles 1000 concurrent users | Non-Functional |

````
# Simple check logic for Functional vs Non-Functional
def verify_scenario(scenario_name, response_time):
    if response_time > 2.0:
        print(f"FAIL: {scenario_name} (Performance issue)")
    else:
        print(f"PASS: {scenario_name}")

verify_scenario("Page Load Time", 1.5)
````

## 3. Professional Test Case Design
A test case is a structured way to verify functionality.

### Test Case Components
*   **Title**: Clear summary.
*   **Preconditions**: Requirements before starting.
*   **Steps**: Clear actions to take.
*   **Expected Result**: The successful outcome.

### Professional Login Scenarios

**TC_LOGIN_001: Valid Login**
* **Steps**: Open browser → Navigate to Login → Enter valid credentials → Click Login.
* **Expected**: Dashboard is displayed.

**TC_LOGIN_002: Invalid Password**
* **Steps**: Enter valid username → Enter wrong password → Click Login.
* **Expected**: Error message displayed.

````
def test_login(username, password):
    # Mock credentials check
    if username == "admin" and password == "1234":
        return "Dashboard Displayed"
    else:
        return "Error: Invalid Credentials"

print(f"Valid Login Result: {test_login('admin', '1234')}")
print(f"Invalid Login Result: {test_login('admin', 'wrong')}")
````

## 5. BDD & Gherkin Syntax

**BDD (Behavior-Driven Development)** focuses on *what* the system should do using simple English.

### The Gherkin Keywords
*   **Given**: Initial condition (Precondition).
*   **When**: Action (User behavior).
*   **Then**: Expected result.
*   **And / But**: Used to extend steps.

### Example Scenario
```gherkin
Scenario: Valid login
  Given user is on login page
  When user enters valid credentials
  And clicks login button
  Then dashboard should be displayed
```

````
# Mocking Gherkin steps in Python
def step_given_on_login():
    print("Navigating to Login Page...")

def step_when_enter_credentials():
    print("Entering username and password...")

step_given_on_login()
step_when_enter_credentials()
````

## 6. TDD vs. BDD Comparison

| Aspect | TDD | BDD |
| :--- | :--- | :--- |
| **Focus** | Code level | User behavior |
| **Language** | Technical | Business-friendly |
| **Audience** | Developers | Entire Team (Dev/QA/BA) |
| **Format** | Unit tests | Given-When-Then |

````
# TDD approach: Write test first
def test_add():
    assert add(2, 3) == 5

def add(a, b):
    return a + b

test_add()
print("TDD Unit Test Passed!")
````

## 7. Professional BDD Scenarios

### Negative Scenario Practice
```gherkin
Scenario: Login failure with incorrect password
  Given user has opened the login page
  When user enters a valid username
  And user enters a wrong password
  And clicks the sign-in button
  Then login should fail
  And error message "Invalid credentials" is shown
```

````
# Handling negative login logic
credentials = {"user": "admin", "pass": "secret"}

def attempt_login(u, p):
    if u != credentials["user"] or p != credentials["pass"]:
        raise ValueError("Invalid credentials")

try:
    attempt_login("admin", "wrong_pass")
except ValueError as e:
    print(f"Caught expected error: {e}")
````

## 8. Gherkin Hierarchy

Structure your features as follows:
1. **Feature**: The high-level requirement.
2. **Scenario**: Specific business case.
3. **Steps**: The sequence of actions.

**Example Feature:**
```gherkin
Feature: Logout functionality
  Scenario: Successful logout
    Given user is logged in
    When user clicks logout
    Then user should be redirected to login page
```

````
# High-level Feature structure simulation
feature = "Logout"
scenarios = ["Successful Logout", "Auto-timeout Logout"]

for s in scenarios:
    print(f"Executing {feature} Scenario: {s}")
````

## 9. Writing Quality BDD Steps

### Golden Rules
*   **Do**: Use business language.
*   **Avoid**: Technical IDs, XPaths, or API endpoints (e.g., `clicks id='btn-01'`).
*   **Specific**: Clearly define the `Then` step.

| Bad Step | Good Step | Reason |
| :--- | :--- | :--- |
| `When user clicks id="login-btn"` | `When user clicks login button` | Remove technical detail |
| `Then it should work` | `Then dashboard should be displayed` | Be specific |

````
# Avoiding technical IDs in logic
def click_button(label):
    print(f"User clicked the '{label}' button")

# Good BDD practice: focus on the action, not the implementation
click_button("Login")
````

## 10. Using 'And' and 'But' for Readability

**'And'** adds similar steps; **'But'** handles exceptions or negative results to make the flow more natural.

**Improved Example:**
```gherkin
Scenario: Login failure
  Given user is on login page
  When user enters valid username
  And user enters wrong password
  And clicks login button
  Then login should fail
  But user should not see dashboard
```

````
# Logic demonstrating 'Then' and 'But'
is_logged_in = False
show_dashboard = False

if not is_logged_in:
    print("Then: Login Failed")
    if not show_dashboard:
        print("But: SUCCESS - Dashboard not visible")
````

## 11. Data-Driven Testing: Scenario Outlines

Avoid duplication by using a single scenario for multiple data sets using the `Examples` table.

```gherkin
Scenario Outline: WMS Login
  Given user is on WMS login page
  When user enters username "<username>"
  And user enters password "<password>"
  Then login should be "<result>"

Examples:
  | username | password | result |
  | user_01  | Pass123  | success |
  | user_01  | wrong    | failure |
```

````
# Simulating a Scenario Outline table with a loop
examples = [
    {'user': 'user_01', 'pass': 'Pass123', 'expected': 'success'},
    {'user': 'user_01', 'pass': 'wrong', 'expected': 'failure'}
]

for data in examples:
    print(f"Testing {data['user']} with {data['pass']} -> Expecting {data['expected']}")
````

## 12. Reusable Setup: Background

Use `Background` to define steps that run before **every** scenario in a feature file to keep your tests DRY (Don't Repeat Yourself).

