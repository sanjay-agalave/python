What is BDD (Simple)

    BDD = Behavior-Driven Development

It focuses on:

    What the system should do ✅
    NOT how it is implemented ❌

Written in simple English so both business + QA + dev understand.

    Why BDD is Important
    Clear requirements
    Easy communication
    Directly usable for automation
    Reduces misunderstandings

Gherkin Syntax (Core of BDD)

    BDD uses Gherkin language

Structure:

    Given → Initial condition
    When → Action
    Then → Expected result

Example (Login)
Normal test case:
    
    Open login page
    Enter username
    Click login
    Verify dashboard

Scenario: Valid login

    Given user is on login page
    When user enters valid username and password
    And user clicks login button
    Then dashboard should be displayed

Understanding Given–When–Then

    Given → Precondition
    What is already true

Examples:

    Given user is on login page
    Given user has valid credentials


When → Action

    What user does

Examples:

    When user enters username
    When user clicks login

Then → Result

    What should happen

Examples:

    Then dashboard is displayed
    Then error message is shown

Use of And

Used for readability (same as Given/When/Then)

    Given user is on login page
    And user has valid account
    When user enters username
    And user enters password
    Then dashboard should be displayed

What is Scenario Outline?

Instead of writing:

Scenario: Login with admin

  Given user is on login page

  When user enters "admin" and "admin123"

  Then login should be successful




Scenario: Login with user1

  Given user is on login page

  When user enters "user1" and "user123"

  Then login should be successful

You can write a single reusable scenario.




Scenario Outline Syntax

Scenario Outline: Login functionality

  Given user is on login page

  When user enters "<username>"

  And user enters "<password>"

  Then login result should be "<result>"




Examples:

  | username | password | result  |

  | admin    | admin123 | success |

  | user1    | user123  | success |




How It Works

Behave will execute:

Execution 1

Given user is on login page

When user enters "admin"

And user enters "admin123"

Then login result should be "success"



Execution 2

Given user is on login page

When user enters "user1"

And user enters "user123"

Then login result should be "success"

Step Definition Example (Python)

@when('user enters username "{username}"')

def step_impl(context, username):

    print(username)

@when('user enters password "{password}"')

def step_impl(context, password):

    print(password)

@then('login result should be "{result}"')

def step_impl(context, result):

    print(result)

Step Definitions Remain Same

You don't need new Python code.

Example:

@given('user launches browser')

    pass




@when('user enters valid username and password')

    pass

Background simply reuses existing steps.




When to Use Background

Use Background when:

Login is required for all scenarios

Common navigation is required

Same preconditions are repeated
